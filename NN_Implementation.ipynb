## **Introduccion**

La gestion de redes se ha vuelto un reto en los ultimos tiempos, debido a la enorme escalabilidad que han tenido las redes en estas ultimas decadas. La gestión de redes, en su concepción inicial, se centraba en tareas relativamente simples, como la supervisión y administración de dispositivos de red para garantizar su funcionamiento óptimo. Sin embargo, a medida que la tecnología de redes ha evolucionado, esta gestión se ha vuelto cada vez más compleja y crítica para las organizaciones.

En sus primeros días, la gestión de redes se basaba principalmente en protocolos y herramientas básicas para monitorear y mantener la infraestructura de red. Las redes eran en su mayoría estáticas y conformadas por un número limitado de dispositivos. La supervisión se centraba en verificar la conectividad y la disponibilidad de los componentes de red, como routers y switches.

Con el tiempo, las redes se han vuelto más dinámicas y abiertas, impulsadas por tendencias como la virtualización, la nube, el Internet de las cosas (IoT) y el 5G. Esto ha llevado a un aumento significativo en la complejidad de la gestión de redes. Algunos de los desafíos actuales incluyen la escalabilidad, la seguridad, la calidad de servicio (QoS), la ubicuidad, el ambiente tecnologico heterogeneo, etc.

Entre todos estos parametros de desempeño hay uno en particular que es de mucha relevancia y es el parametro del delay o retraso en la entrega de los paquetes en una red. Para la mayoria de tendencias actuales es de suma importancia que el delay sea muy pequeño, lo mas que se pueda, para que las redes que manejan estas tecnologias puedan funcionar de manera adecuada. En este problema se abordará como predecir el delay de una conexion extremo a extremo en base a una "snapshot" de la red haciendo uso del ML, el cual es fundamental para tareas de optimizacion y gestion de redes en la actualidad.

## **Marco Teórico**

1. **Redes neuronales**

Las redes neuronales son modelos simples del funcionamiento del sistema nervioso. Las unidades básicas son las neuronas, que generalmente se organizan en capas, como se muestra en la siguiente ilustración.

image.png

Una red neuronal es un modelo simplificado que emula el modo en que el cerebro humano procesa la información: Funciona simultaneando un número elevado de unidades de procesamiento interconectadas que parecen versiones abstractas de neuronas.

Las unidades de procesamiento se organizan en capas. Hay tres partes normalmente en una red neuronal : una capa de entrada, con unidades que representan los campos de entrada; una o varias capas ocultas; y una capa de salida, con una unidad o unidades que representa el campo o los campos de destino. Las unidades se conectan con fuerzas de conexión variables (o ponderaciones). Los datos de entrada se presentan en la primera capa, y los valores se propagan desde cada neurona hasta cada neurona de la capa siguiente. al final, se envía un resultado desde la capa de salida.

La red aprende examinando los registros individuales, generando una predicción para cada registro y realizando ajustes a las ponderaciones cuando realiza una predicción incorrecta. Este proceso se repite muchas veces y la red sigue mejorando sus predicciones hasta haber alcanzado uno o varios criterios de parada.



2. **Redes neuronales de grafos (Graph Neural Networks)**

Estas redes son un tipo especial de red neuronal que funciona directamente sobre grafos a diferencia de las redes neuronales típicas reciben de entrada información codificada como vectores, matrices o tensores. Esta diferencia, permite que las GNN, puedan aprender a partir de información más compleja, ya que no son sólo números aislados, si no que se incorporan las relaciones entre estos.

Un grafo es una forma de organizar la información como nodos o vértices que representan algún dato y relaciones, llamadas aristas, entre estos. Debido a esto, pueden ser más complejas de analizar que imágenes o texto. Por esa razón las redes neuronales típicas no funcionan de forma adecuada.

Las GNNs se inspiran en las redes neuronales convolucionales (CNN) y recurrentes (RNN), pero están diseñadas para trabajar con datos de grafos en lugar de datos tabulares o secuenciales. Los componentes clave de las redes neuronales de grafos:



*   Nodos y características: Cada nodo en el grafo representa una entidad o elemento de datos, y cada nodo puede tener asociadas características específicas. Por ejemplo, en una red social, los nodos podrían ser usuarios, y las características podrían incluir información sobre sus intereses o conexiones con otros usuarios.
*   Conexiones y relaciones: Las aristas en el grafo representan las relaciones o conexiones entre nodos. Estas relaciones pueden ser dirigidas o no dirigidas y pueden llevar información adicional, como la fuerza de la conexión o el tipo de relación.
*   Agrupación de vecinos: Una de las características distintivas de las GNNs es su capacidad para considerar la información de los vecinos de un nodo. A medida que propagan información a través del grafo, las GNNs agregan y combinan las características de los nodos vecinos para enriquecer la representación del nodo actual.
*   Capas de aprendizaje: Las GNNs constan de múltiples capas, cada una de las cuales actualiza la representación de los nodos. Estas capas pueden ser similares a las capas en otros tipos de redes neuronales, como las capas de convolución en las CNN o las capas recurrentes en las RNN.

3. **Retardo de paquetes**


El "per-source-destination mean per-packet delay" es un término que se utiliza en el contexto de las redes de comunicación para medir el retardo promedio por paquete desde una fuente a un destino específico.

Este concepto es importante en la evaluación del rendimiento de una red, ya que el retardo de extremo a extremo es un factor crítico en la calidad de servicio (QoS) de muchas aplicaciones de red. Un retardo excesivo puede llevar a problemas como la pérdida de paquetes, la degradación de la calidad de voz y video en aplicaciones en tiempo real, y una experiencia de usuario deficiente en general.

El "per-source-destination mean per-packet delay" se calcula generalmente tomando múltiples mediciones de retardo de paquetes individuales enviados desde una fuente particular a un destino particular durante un período de tiempo. Luego, se calcula el promedio de estos valores para obtener el retardo promedio. Este valor puede variar según la congestión de la red, la distancia entre la fuente y el destino, la ruta que sigue el paquete y otros factores relacionados con la topología y el rendimiento de la red.

Un retardo teorico se calcularía haciendo uso de la siguiente ecuación:

$T=\frac{d}{v}$

donde $d$ es la distancia del enlace y $v$ es la velocidad con la que viaja el paquete.

Hay que tener en cuenta otro retardo: el retardo de cola, este es el tiempo que permanece un conjunto de datos en la cola de espera hasta ser definitivamente reenviado.

Los retardos de cola también se denominan de almacenamiento ya que es el tiempo, una vez que han sido procesados, los datos permanecen en el nodo. Este tipo de retardo generalmente se debe debido a que existe un mayor número de paquetes que entran de los que salen. Normalmente, estos dispositivos funcionan como cola FIFO, es decir, el primero en entrar es el primero en salir. Aunque se pueden implementar otro tipo de lógicas en función de la prioridad que tengan los datos.

Para reducir el retardo de cola se debe aumentar el caudal de salida con el fin de disminuir el “atasco” en las puertas de salida del nodo. Aunque también se pueden implementar distintas políticas en la red relacionadas con limitar el número de datos que llegan al nodo.

Para el cálculo analítico de este retardo hay que irse al aparato matemático de la teoría de colas, el cual, fundamentalmente, se basa en un tratamiento estadístico, dado que para el caso que nos ocupa, los paquetes de datos no tienen todos el mismo tamaño y pueden llegar a ráfagas (es decir, no periódicamente). Por tanto, su cálculo es estadístico. Aunque para su manejo se pueden hacer una serie de consideraciones prácticas.

El conjunto de datos de entrenamiento contiene solo muestras simuladas en las topologías de red NSFNET y GEANT2, mientras que el conjunto de datos de validación contiene muestras simuladas en GBN.


En particular, el conjunto de datos se centra en agregar políticas de programación de colas, donde cada nodo se configura con diferentes políticas de programación de acuerdo con los escenarios que se explican a continuación. Se utilizan las siguientes políticas: Prioridad estricta (SP), donde los paquetes en colas con más prioridad se transmiten primero. Cola justa ponderada (WFQ) y Deficit Round Robin (DRR).

- Strict priority (SP): Puede configurar una cola por interfaz para que tenga prioridad estricta, lo que hace que el tráfico sensible al retraso, como el tráfico de voz, se elimine y reenvíe con un retraso mínimo. Los paquetes que están en cola en una cola de prioridad estricta se eliminan antes que los paquetes en otras colas, incluidas las colas de alta prioridad.
La función de colas estrictas de alta prioridad le permite configurar la vigilancia del tráfico que evita que las colas de menor prioridad se queden sin acceso.
- Weighted Fair Queueing (WFQ): Esta estrategia consta de implementaciones de una política generalizada de uso compartido de procesadores (GPS) y una generalización natural de colas justas (FQ). WFQ permite que cada flujo tenga una cierta proporción de capacidad de enlace, que generalmente está especificada por el propio flujo.
- Deficit Round Robin (DRR): intenta proporcionar ciertas garantías de ancho de banda mediante el uso de un acumulador de déficit para determinar si un paquete se puede transferir durante un ciclo del programador. Se agrega una cantidad de cuantos al acumulador de déficit de cada cola durante cada ciclo del programador. Si el tamaño del paquete pendiente es menor que el valor del acumulador, el paquete es elegible para transferencia.

Para todas las políticas de programación de colas implementadas, siempre consideramos tres colas en el puerto de salida de cada nodo. Así, cada muestra del conjunto de datos incluye una matriz de tráfico donde los flujos pueden tener 3 Tipos de Servicios diferentes (ToS=[0,1,2]) respectivamente asociados a una de las tres colas con diferente prioridad (por ejemplo, ToS=0 está asociado a la primera cola de nodos). Al inicio de cada simulación, se asocia un ToS a los flujos generados en un camino origen-destino. Todos los paquetes de la ruta tendrán el mismo ToS. Los paquetes se generan en cada flujo siguiendo una distribución de Poisson. Con este fin, se utiliza una distribución de tiempo exponencial para modelar los tiempos de llegada entre paquetes. Asimismo, se utiliza una distribución binomial para modelar el tamaño del paquete. El bitrate máximo que pueden tener los caminos se selecciona aleatoriamente para cada simulación, entre 400 y 2000 bits por unidad de tiempo. Tenga en cuenta que el retraso de los paquetes está limitado a 20 unidades de tiempo para todos los escenarios de simulación.

Con este criterio se generan un total de 100 configuraciones de programación (es decir, nodos + perfiles de peso) para cada topología, y cada simulación selecciona aleatoriamente una de estas configuraciones:

## Procesamiento inicial del dataset:
Se utiliza una API proporcionada por el desafío para un tratamiento mas sencillo de los datos del dataset por medio de funciones sencillas para extraer las caracteristicas almacenadas. El objeto principal de esta API es el objeto llamado *sample*. Cada *sample* es un contenedor que incluye información sobre: ​​(i) una topología de red, (ii) parámetros utilizados para generar el tráfico en nuestra herramienta de simulación, (iii) una configuración de enrutamiento y (iv) algunas medidas (retraso, jitter y pérdida). medido con la simulación resultante de la configuración de la red (es decir, topología + distribución de tráfico + enrutamiento). Tener en cuenta que se considera solo una ruta que conecta cada par fuente-destino (src-dst); sin embargo, más de un flujo src-dst puede atravesar la misma ruta y las mediciones de rendimiento se pueden obtener individualmente para cada flujo. Cada instancia de un *sample* comprende los siguientes atributos:
- global_packets: Número de paquetes transmitidos en la red por unidad de tiempo (paquetes/unidad de tiempo).
- global_losses: Paquetes perdidos en la red por unidad de tiempo (paquetes/unidad de tiempo).
- global_delay: Retraso promedio por paquete sobre todos los paquetes transmitidos en la red (unidades de tiempo).
- maxAvgLambda:
es la tasa de tráfico promedio máxima (bits/unidad de tiempo) que una ruta puede generar en el escenario de simulación.
- performance_matrix:
Matriz con medidas agregadas de rendimiento de src-dst y nivel de flujo (retraso, jitter y pérdida) medidas en cada par src-dst.
- traffic_matrix: Matriz con las distribuciones de tiempo y tamaño utilizadas para generar tráfico para cada par src-dst.
- routing_matrix:
Matriz con las rutas para conectar cada par src-dst.
- topology_object:
Utiliza un objeto Graph de la biblioteca Networkx que incluye información relacionada con la topología a nivel de nodo y enlace.

Ahora se verá mas en detalle la matriz de desempeño y la matriz de tráfico para entender la estructura de los datos.

**Matriz de desempeño**: Esta es una matriz que indexa las mediciones de rendimiento a nivel de pares src-dst. En particular, considera que se puede intercambiar más de un flujo en cada par src-dst. Por lo tanto, proporciona mediciones de rendimiento en dos niveles de granularidad: (i) para todos los flujos agregados en cada par src-dst, y (ii) para cada flujo individualmente. Cada elemento de esta matriz contiene un diccionario con las siguientes claves: *'AggInfo'* (diccionario con medidas de rendimiento para todos los flujos agregados entre un par [src,dst] específico. Los siguientes son algunos de los parametros que contiene: [‘PktsDrop’, ‘AvgDelay’, ‘Jitter’]) y *'Flows'* (lista de diccionarios con medidas de rendimiento para cada flujo entre el nodo i y el nodo j. Los flujos en cada par src-dst están indexados mediante una identificación numérica. Los siguientes son algunos de los parametros que contiene: [‘PktsDrop’, ‘AvgDelay’, ‘Jitter’]).

**Matriz de tráfico**: Esta matriz indexa información relacionada con el tráfico a nivel de pares src-dst. De manera similar a performance_matrix, proporciona información de tráfico en dos niveles de granularidad: (i) para todos los flujos agregados en cada par src-dst y (ii) para cada flujo individualmente. Cada elemento de esta matriz (es decir, Traffic_matrix[src,dst]) contiene un diccionario con las siguientes claves: *'AggInfo'* (diccionario con mediciones de tráfico para todos los flujos agregados entre un par [src,dst] específico. Los siguientes son los parametros que contiene: [‘AvgBw’,‘PktsGen’]), y *'Flows'* (lista de diccionarios con información de tráfico a nivel de flujo para cada flujo entre el nodo src y el nodo dst. Incluye las mediciones de tráfico para cada flujo y los parámetros utilizados para generar el flujo de tráfico en el simulador de red. Los siguientes son los parametros que contiene: [‘TimeDis’,‘TimeDistParams,‘SizeDist’,‘SizeDistParams’,‘AvgBw’,‘PktsGen’,‘ToS’])

**Matriz de enrutamiento**:
Esta matriz describe la configuración de enrutamiento. En particular, incluye todas las rutas que conectan cada par src-dst. En particular, esta lista incluye los ID de los nodos que atraviesa la ruta. Tenga en cuenta que todos los flujos de un par src-dst siguen el mismo camino.

**Topologia**: Este es un objeto Graph de la biblioteca Networkx que proporciona información sobre la topología de la red. En particular, este objeto codifica información sobre nodos y enlaces. A traves de este objeto se puede acceder a las politicas de programacion de colas escogida por cada nodo, los puertos, el ancho de banda, entre otros.

## Visualizar y procesar el dataset TRAINNING

```python
import datanetAPI
import matplotlib.pyplot as plt
import numpy as np

reader = datanetAPI.DatanetAPI("/data/sample_data/train",[], [0])  # se lee el dataset de trainning
it = iter(reader)
n = 400   # numero total de objetos sample para el dataset de entrenamiento = 400

# listas para almacenar datos de interes (hay 2 topologias y se separan respectivamente)
average_bandwidth_top1 = []
total_packets_top1 = []
average_bandwidth_top2 = []
total_packets_top2 = []
average_delay_top1 = []
drops_top1 = []
jitter_top1 = []
average_delay_top2 = []
drops_top2 = []
jitter_top2 = []
path_top1 = []
param1_top1 = []
param2_top1 = []
path_top2 = []
param1_top2 = []
param2_top2 = []
global_delay_top1 = []
global_delay_top2 = []
global_loss_top1 = []
global_loss_top2 = []
global_packets_top1 = []
global_packets_top2 = []

# ejemplo del enlace 0-1
src = 0
dst = 1

for i in range(n):
    sample = next(it)
    m_tr = sample.get_traffic_matrix()   # se extrae la matriz de trafico
    tamaño = sample.get_network_size()  # Cantidad de nodos en la red de la muestra
    paquets = m_tr[src,dst]['AggInfo']['TotalPktsGen']  # Paquetes totales generados en todos los flujos para el enlace 0-1 de la muestra
    bandwidth = m_tr[src,dst]['AggInfo']['AvgBw']  # Ancho de banda promedio en todos los flujos para el enlace 0-1 de la muestra
    if tamaño == 24:  # topologia 1 GEANT2
      average_bandwidth_top1.append(bandwidth)
      total_packets_top1.append(paquets)
    elif tamaño == 14:  # topologia 2 NSFNET
      average_bandwidth_top2.append(bandwidth)
      total_packets_top2.append(paquets)

    m_per = sample.get_performance_matrix()  # se extrae la matriz de desempeño
    delay = m_per[src,dst]['AggInfo']['AvgDelay']  # delay promedio de todos los flujos en el enlace 0-1
    drop = m_per[src,dst]['AggInfo']['PktsDrop']  # paquetes descartados por unidad de tiempo en todos los flujos del enlace 0-1
    jitter = m_per[src,dst]['AggInfo']['Jitter']  # variacion del delay por paquete de los paquetes transmitidos en todos los flujos del enlace 0-1
    if tamaño == 24:  # topologia 1 GEANT2
      average_delay_top1.append(delay)
      drops_top1.append(drop)
      jitter_top1.append(jitter)
    elif tamaño == 14:  # topologia 2 NSFNET
      average_delay_top2.append(delay)
      drops_top2.append(drop)
      jitter_top2.append(jitter)

    m_route = sample.get_routing_matrix()   # se extrae la matriz de enrutamiento
    path = m_route[src,dst]   #  lista que describe el path entre 0 y 1
    if tamaño == 24:  # topologia 1 GEANT2
      path_top1.append(path)
    elif tamaño == 14:  # topologia 2 NSFNET
      path_top2.append(path)

    topology = sample.get_topology_object()  # se extrae el objeto topologia
    param1 = topology.nodes[src]  # entrega el nivel de QoS, el tamaño de cola, la politica de programacion de cola que se escogio y los pesos del nodo 0
    param2 = topology.nodes[dst]  # entrega el nivel de QoS, el tamaño de cola, la politica de programacion de cola que se escogio y los pesos del nodo 1
    if tamaño == 24:  # topologia 1 GEANT2
      param1_top1.append(param1)
      param2_top1.append(param2)
    elif tamaño == 14:  # topologia 2 NSFNET
      param1_top2.append(param1)
      param2_top2.append(param2)


    gdelay = sample.get_global_delay()   # se extrae el delay global de la red
    gloss = sample.get_global_losses()  # se extraen los paquetes perdidos en la red
    gpacket = sample.get_global_packets()  # se extraen los paquetes generados en toda la red
    if tamaño == 24:  # topologia 1 GEANT2
      global_delay_top1.append(gdelay)
      global_loss_top1.append(gloss)
      global_packets_top1.append(gpacket)
    elif tamaño == 14:  # topologia 2 NSFNET
      global_delay_top2.append(gdelay)
      global_loss_top2.append(gloss)
      global_packets_top2.append(gpacket)
```

## Visualizar y procesar el dataset VALIDATION

```python
reader = datanetAPI.DatanetAPI("/data/sample_data/validation",[], [0])  # se lee el dataset de validation
it = iter(reader)
n = 200   # numero total de objetos sample para el dataset de validacion = 200

# listas para almacenar datos de interes (solo hay una topologia)
average_bandwidth = []
total_packets = []
average_delay = []
drops = []
jitters = []
paths = []
params1 = []
params2 = []
global_delay = []
global_loss = []
global_packets = []

# ejemplo del enlace 0-10
src = 0
dst = 10

for i in range(n):
    sample = next(it)
    m_tr = sample.get_traffic_matrix()   # se extrae la matriz de trafico
    tamaño = sample.get_network_size()  # Cantidad de nodos en la red de la muestra
    paquets = m_tr[src,dst]['AggInfo']['TotalPktsGen']  # Paquetes totales generados en todos los flujos para el enlace 0-10 de la muestra
    bandwidth = m_tr[src,dst]['AggInfo']['AvgBw']  # Ancho de banda promedio en todos los flujos para el enlace 0-10 de la muestra
    average_bandwidth.append(bandwidth)
    total_packets.append(paquets)

    m_per = sample.get_performance_matrix()  # se extrae la matriz de desempeño
    delay = m_per[src,dst]['AggInfo']['AvgDelay']  # delay promedio de todos los flujos en el enlace 0-10
    drop = m_per[src,dst]['AggInfo']['PktsDrop']  # paquetes descartados por unidad de tiempo en todos los flujos del enlace 0-10
    jitter = m_per[src,dst]['AggInfo']['Jitter']  # variacion del delay por paquete de los paquetes transmitidos en todos los flujos del enlace 0-10
    average_delay.append(delay)
    drops.append(drop)
    jitters.append(jitter)

    m_route = sample.get_routing_matrix()   # se extrae la matriz de enrutamiento
    path = m_route[src,dst]   #  lista que describe el path entre 0 y 10
    paths.append(path)

    topology = sample.get_topology_object()  # se extrae el objeto topologia
    param1 = topology.nodes[src]  # entrega el nivel de QoS, el tamaño de cola, la politica de programacion de cola que se escogio y los pesos del nodo 0
    param2 = topology.nodes[dst]  # entrega el nivel de QoS, el tamaño de cola, la politica de programacion de cola que se escogio y los pesos del nodo 10
    params1.append(param1)
    params2.append(param2)

    gdelay = sample.get_global_delay()   # se extrae el delay global de la red
    gloss = sample.get_global_losses()  # se extraen los paquetes perdidos en la red
    gpacket = sample.get_global_packets()  # se extraen los paquetes generados en toda la red
    global_delay.append(gdelay)
    global_loss.append(gloss)
    global_packets.append(gpacket)
```
